```python
def is_prime_recursive(n , i = 2):

  if n <= 2:
    return True if n == 2 else False

  if n % i == 0:
    return False

  if i * i > n:
    return True

  return is_prime_recursive(n , i + i)


print(is_prime_recursive(25))
```

```python
def is_prime_iterative(n):
  if n <= 1:
    return False

  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False

  return True

print(is_prime_iterative(67))
```

Check if a number and String is a palindrome.

```python
#Using Iteration

text = input("Enter a string: ")
if text == text[::-1]:
    print("It is a palindrome string.")
else:
    print("It is not a palindrome string.")
```

```python
#Using Recursion

def is_palidrome_string(s):
  if len(s) <= 1:
    return True

  if s[0] != s[-1]:
    return False

  return is_palidrome_string(s[1 : -1])

print(is_palidrome_string('Bitch'))
print(is_palidrome_string('MADAM'))
```

```python
num = int(input("Enter the Number : "))
temp = num
rev = 0

while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10

if temp == rev:
  print('its a palidrome number : ')

else:
  print('its not a palidrome number : ')

```

```python
def reverse_num(n, rev = 0):
  if n == 0:
    return rev

  return reverse_num(n // 10, rev * 10 + n % 10)

num = int(input("Enter a Number : "))
if num == reverse_num(num):
  print("Its a palidrome number")

else:
  print('Its not a palidrome number')
```
