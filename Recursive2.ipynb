Find the PRIME NUMBER using RECURSION

```python
def is_prime(n, i = 2):

#BASE CASE
  if n <= 2:
    return True if n == 2 else False

  if n % i == 0:
    return False

  if i * i > n:
    return True

#RECURSIVE CASE
  return is_prime(n , i + 1)

num = 25
print(f"{num} is a prime number = " , is_prime(num))
```

```python
def is_palindrome_string(s):
  if len(s) <= 1:
    return True
  if s[0] != s[-1]:
    return False

  return is_palindrome_string(s[1 : -1])

print(is_palindrome_string("MADAM"))
print(is_palindrome_string("Bitch"))
```

```python
def is_string(s):

  if len(s) <= 1:
    return True
  if s[0] != s[-1]:
    return False

  return is_string(s[ 1 : -1])

print(is_string("MADAM"))
print(is_string("Bitch"))
```

```python
def reverse_string(s: str):

  if s == '':
    return ''

  # Take the last character and prepend it to the reversed rest of the string
  return reverse_string(s[1:]) + s[0]

print(reverse_string('you are a bitch'))
```
