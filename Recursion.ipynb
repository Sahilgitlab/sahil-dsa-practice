Recursion reverse **string**

```python
def reverse_string(s : str):
  if s == '':
    return ''
  return reverse_string(s[1 : ]) + s[0]

print(reverse_string('hello bitch ass nigga'))
```

Recursion reverse List

```python
def reverse_lst(lst):
  if lst == []:
    return []
  return reverse_lst(lst[1 : ]) + [lst[0]]

print(reverse_lst([1,2,3,4,5,6,7,8,9]))
```

fiboncci recursion

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fiboncci(num):
  if num < 2:
    return num
  return fiboncci(num - 1) + fiboncci(num - 2)

print(fiboncci(100))
```

Factorial using recursion

```python
def factorial(num):
  if num == 0 or num == 1:
    return 1

  return num * factorial(num - 1)

print(factorial(15))
```

Power Recursion


```python

def power(base, exponent):
  if exponent == 0:
    return 1

  return base * power(base, exponent - 1 )

print(power(5, 88))
```

Array sum using recursion

```python
def array_sum(arr):
  if len(arr) == 0:
    return 0

  else:
    return arr[0] + array_sum(arr[1 : ])


print(array_sum([1, 2, 3, 4, 5]))
```
